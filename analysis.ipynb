```python
import pandas as pd
import ast

# Define the sets of positive and negative words
turkish_positive = set(turkish_positive_text.replace("\n", "").split(", "))
turkish_negative = set(turkish_negative_text.replace("\n", "").split(", "))

spanish_positive = set(spanish_positive_text.replace("\n", "").split(", "))
spanish_negative = set(spanish_negative_text.replace("\n", "").split(", "))

dutch_positive = set(dutch_positive_text.replace("\n", "").split(", "))
dutch_negative = set(dutch_negative_text.replace("\n", "").split(", "))

def classify_words_from_transcript(transcript, positive_set, negative_set):
    """
    Classify words from the 'clean_transcript' column into positive and negative categories.
    
    Parameters:
        transcript (str): The text to classify.
        positive_set (set): Set of positive words.
        negative_set (set): Set of negative words.
    
    Returns:
        tuple: Two lists containing positive and negative words.
    """
    positive = []
    negative = []
    
    if isinstance(transcript, str):  # Ensure the input is a string
        for word in transcript.split():
            word = word.strip().lower()  # Normalize the word
            if word in positive_set:
                positive.append(word)
            elif word in negative_set:
                negative.append(word)
    
    return positive, negative

# Apply the classification function to each DataFrame
turkish_df[['MENTAL_HEALTH_POSITIVE', 'MENTAL_HEALTH_NEGATIVE']] = turkish_df['clean_transcript'].apply(
    lambda x: pd.Series(classify_words_from_transcript(x, turkish_positive, turkish_negative))
)

spanish_df[['MENTAL_HEALTH_POSITIVE', 'MENTAL_HEALTH_NEGATIVE']] = spanish_df['clean_transcript'].apply(
    lambda x: pd.Series(classify_words_from_transcript(x, spanish_positive, spanish_negative))
)

dutch_df[['MENTAL_HEALTH_POSITIVE', 'MENTAL_HEALTH_NEGATIVE']] = dutch_df['clean_transcript'].apply(
    lambda x: pd.Series(classify_words_from_transcript(x, dutch_positive, dutch_negative))
)

# Display the DataFrames for verification
display(turkish_df)
display(spanish_df)
display(dutch_df)
```

```python
import matplotlib.pyplot as plt

def sum_combined_columns(pronoun_df, verb_columns, pronoun_columns, include_verbs=True):
    """
    Sum pronouns and verbs into unified categories. For languages without verb columns (dutch), 
    only pronouns are processed.
    """
    combined_sums = {}
    for pronoun_col in pronoun_columns:
        # Sum pronoun counts
        pronoun_sum = pronoun_df[pronoun_col].sum() if pronoun_col in pronoun_df.columns else 0
        # If verbs are included, sum their counts
        if include_verbs:
            verb_col = pronoun_col + "_verbs"
            verb_sum = (
                sum(len(items) if isinstance(items, list) else 0 for items in pronoun_df[verb_col])
                if verb_col in pronoun_df.columns
                else 0
            )
            combined_sums[pronoun_col] = pronoun_sum + verb_sum
        else:
            combined_sums[pronoun_col] = pronoun_sum
    return combined_sums

# Define the pronoun categories
interest_columns_pronouns = ["FPS", "FPP", "SPS", "SPP", "TPS", "TPP"]

# Calculate combined sums for Spanish and Turkish (pronouns + verbs) and Dutch (pronouns only)
spanish_combined_sum = sum_combined_columns(spanish_pronoun_df, interest_columns_pronouns, interest_columns_pronouns, include_verbs=True)
turkish_combined_sum = sum_combined_columns(turkish_pronoun_df, interest_columns_pronouns, interest_columns_pronouns, include_verbs=True)
dutch_combined_sum = sum_combined_columns(dutch_pronoun_df, interest_columns_pronouns, interest_columns_pronouns, include_verbs=False)

# Normalize the sums for visualization
spanish_total = sum(spanish_combined_sum.values())
turkish_total = sum(turkish_combined_sum.values())
dutch_total = sum(dutch_combined_sum.values())

spanish_normalized = {k: v / spanish_total * 100 for k, v in spanish_combined_sum.items()}
turkish_normalized = {k: v / turkish_total * 100 for k, v in turkish_combined_sum.items()}
dutch_normalized = {k: v / dutch_total * 100 for k, v in dutch_combined_sum.items()}

# Prepare data for plotting
categories = list(spanish_normalized.keys())
spanish_values = [spanish_normalized[category] for category in categories]
turkish_values = [turkish_normalized[category] for category in categories]
dutch_values = [dutch_normalized[category] for category in categories]

x = range(len(categories))

# Create the plot
plt.figure(figsize=(12, 6))

plt.bar([i - 0.3 for i in x], spanish_values, width=0.25, label='Spanish', color='blue', align='center')
plt.bar(x, turkish_values, width=0.25, label='Turkish', color='red', align='center')
plt.bar([i + 0.3 for i in x], dutch_values, width=0.25, label='Dutch', color='green', align='center')

# Add labels and legend
plt.xlabel('Combined Categories (Pronouns + Verbs)')
plt.ylabel('Percentage (%)')
plt.title('Normalized Pronoun Usage Comparison (Spanish, Turkish, Dutch)')
plt.xticks(x, categories)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Assuming the data is in DataFrames: turkish_df, spanish_df, and dutch_df
# Each contains the columns "MENTAL_HEALTH_POSITIVE" and "MENTAL_HEALTH_NEGATIVE"

from collections import Counter
import matplotlib.pyplot as plt

# Function to aggregate word counts from the DataFrame column
def aggregate_word_counts(df, column):
    words = [word for sublist in df[column] for word in sublist]
    return Counter(words)

# Aggregate counts for each language
spanish_counts = {
    "MENTAL_HEALTH_POSITIVE": aggregate_word_counts(spanish_df, "MENTAL_HEALTH_POSITIVE"),
    "MENTAL_HEALTH_NEGATIVE": aggregate_word_counts(spanish_df, "MENTAL_HEALTH_NEGATIVE"),
}

turkish_counts = {
    "MENTAL_HEALTH_POSITIVE": aggregate_word_counts(turkish_df, "MENTAL_HEALTH_POSITIVE"),
    "MENTAL_HEALTH_NEGATIVE": aggregate_word_counts(turkish_df, "MENTAL_HEALTH_NEGATIVE"),
}

dutch_counts = {
    "MENTAL_HEALTH_POSITIVE": aggregate_word_counts(dutch_df, "MENTAL_HEALTH_POSITIVE"),
    "MENTAL_HEALTH_NEGATIVE": aggregate_word_counts(dutch_df, "MENTAL_HEALTH_NEGATIVE"),
}

# Normalize counts
def normalize_counts(counts):
    total = sum(sum(cat.values()) for cat in counts.values())
    if total > 0:
        return {k: sum(v.values()) / total * 100 for k, v in counts.items()}
    else:
        return {k: 0 for k in counts.keys()}

spanish_normalized = normalize_counts(spanish_counts)
turkish_normalized = normalize_counts(turkish_counts)
dutch_normalized = normalize_counts(dutch_counts)

# Prepare data for plotting
categories = ["MENTAL_HEALTH_POSITIVE", "MENTAL_HEALTH_NEGATIVE"]

spanish_values = [spanish_normalized[category] for category in categories]
turkish_values = [turkish_normalized[category] for category in categories]
dutch_values = [dutch_normalized[category] for category in categories]

x = range(len(categories))

# Create the plot
plt.figure(figsize=(10, 6))

plt.bar(x, spanish_values, width=0.25, label='Spanish', color='lightpink', align='center')
plt.bar([i + 0.3 for i in x], turkish_values, width=0.25, label='Turkish', color='lightgreen', align='center')
plt.bar([i - 0.3 for i in x], dutch_values, width=0.25, label='Dutch', color='lightblue', align='center')

# Add labels and legend
plt.xlabel('Mental Health Categories')
plt.ylabel('Percentage (%)')
plt.title('Normalized Mental Health Word Volume Comparison by Language')
plt.xticks(x, categories)
plt.legend()

# Display the plot
plt.show()
```

```python
def compare_word_counts(df):
    """
    Compute total word count from 'clean_transcript' column and compare it with
    the counts of 'MENTAL_HEALTH_POSITIVE' and 'MENTAL_HEALTH_NEGATIVE'.
    Dynamically adjusts output key to include the language in 'Transcript_' column.
    """
    # Detect language from the first column that starts with 'Transcript_'
    language_col = [col for col in df.columns if col.startswith("Transcript_")][0]
    language = language_col.split("_")[1]  # Extract language from column name

    # Ensure 'clean_transcript' is processed as a string and handle NaN values
    total_word_count = df['clean_transcript'].apply(lambda x: len(str(x).split()) if isinstance(x, str) else 0).sum()
    
    # Count positive and negative mental health mentions
    positive_mention_count = df['MENTAL_HEALTH_POSITIVE'].apply(lambda x: len(x) if isinstance(x, list) else 0).sum()
    negative_mention_count = df['MENTAL_HEALTH_NEGATIVE'].apply(lambda x: len(x) if isinstance(x, list) else 0).sum()
    
    # Prepare comparison summary
    comparison = {
        f"Total Words in {language} transcript": total_word_count,
        "Total Mental Health Positive Mentions": positive_mention_count,
        "Total Mental Health Negative Mentions": negative_mention_count,
        "Positive to Total Words (%)": (positive_mention_count / total_word_count) * 100 if total_word_count > 0 else 0,
        "Negative to Total Words (%)": (negative_mention_count / total_word_count) * 100 if total_word_count > 0 else 0,
    }
    
    # Return the comparison results
    return comparison

comparison_results_spanish = compare_word_counts(spanish_df)
print(comparison_results_spanish)

comparison_results_dutch = compare_word_counts(dutch_df)
print(comparison_results_dutch)

comparison_results_turkish = compare_word_counts(turkish_df)
print(comparison_results_turkish)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter

def generate_word_cloud(df, column_names, title, colormap):
    """
    Generate a word cloud from specified columns in a DataFrame.
    
    Parameters:
        df (DataFrame): The DataFrame containing the columns.
        column_names (list): List of column names to extract the words from.
        title (str): Title for the word cloud plot.
    """
    # Combine all words from the specified columns
    words = []
    for column in column_names:
        if column in df.columns:
            words += df[column].explode().dropna().tolist()
    
    # Count the word frequencies
    word_counts = Counter(words)
    
    # Generate the word cloud
    wordcloud = WordCloud(width=800, height=400, background_color="white", colormap=colormap).generate_from_frequencies(word_counts)
    
    # Plot the word cloud
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title(title, fontsize=16)
    plt.show()

# Specify the mental health columns
mental_health_columns = ["MENTAL_HEALTH_POSITIVE", "MENTAL_HEALTH_NEGATIVE"]

# Generate and display the word cloud
generate_word_cloud(spanish_df, mental_health_columns, title="Mental Health Word Cloud Spanish", colormap="Blues_r")
generate_word_cloud(turkish_df, mental_health_columns, title="Mental Health Word Cloud Turkish", colormap="Reds_r")
generate_word_cloud(dutch_df, mental_health_columns, title="Mental Health Word Cloud Dutch", colormap="Greens")
```
